```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df0 = pd.read_csv('course_grades.csv',sep=';',decimal=',')
df = df0.iloc[:,-5:]
df
```

```python

plt.scatter(df.exercises, df.STACKexer)


plt.xlabel('exercises', size =20)
plt.ylabel('STACKexer',  size =20)
plt.show()
```

```python
df_exercises_STACKexer = df[['exercises','STACKexer']]
df_exercises_STACKexer.cov() 
```

```python
df_exercises_STACKexer.corr() 
# positive covariance and Correlation =  0.85 indicates a relationship between exercises & STACKexer  
# the plot indications are similar to calculations
```

```python
plt.scatter(df.exam1, df.exam2)


plt.xlabel('exam1', size =20)
plt.ylabel('exam2',  size =20)
plt.show()
```

```python
df_exam1_exam2 = df[['exam1','exam2']]
df_exam1_exam2.cov() # A higher covariance between exam1 and exam2 (strong relationship)
```

```python
df_exam1_exam2.corr()
```

```python
df_exam1_grade=df[['exam1', 'grade']]
df_exam2_grade = df[['exam2', 'grade']]
plt.figure(figsize=(10,3))
plt.subplot(1, 2, 1)

plt.scatter(df_exam1_grade.grade, df_exam1_grade.exam1)
plt.xlabel('grade', size =20)
plt.ylabel('exam1',  size =20)


plt.subplot(1, 2, 2)

plt.scatter(df_exam2_grade.grade, df_exam2_grade.exam2)
plt.xlabel('grade', size =20)
plt.ylabel('exam2',  size =20)
plt.show()

#The figure below show that there is a relationship between grades and the exams because as exams grow the grade grows 
```

```python
print("Exam1 & grade Covariance:   -------------------------  " )
print(df_exam1_grade.cov())
print("\n\nExam1 & grade Correlation:   ------------------------- ")
print(df_exam1_grade.corr())
print("\n\nExam2 & grade Covariance:   -------------------------  " )
print(df_exam2_grade.cov())
print("\n\nExam2 & grade Correlation:   ------------------------- ")
print(df_exam2_grade.corr())
```
